```python
# Advent of Code 2015, Day 21
from itertools import combinations

# Constants
l_weapons = [("Dagger", 8, 4, 0),
             ("Shortsword", 10, 5, 0), 
             ("Warhammer", 25, 6, 0),
             ("Longsword", 40, 7, 0),
             ("Greataxe", 74, 8, 0)]
l_armor = [("Leather", 13, 0, 1),
           ("Chainmail", 31, 0, 2),
           ("Splintmail", 53, 0, 3),
           ("Bandedmail", 75, 0, 4),
           ("Platemail", 102, 0, 5),
           ("No Armor", 0, 0, 0)]
l_rings = [("Damage +1", 25, 1, 0),
           ("Damage +2", 50, 2, 0),
           ("Damage +3", 100, 3, 0),
           ("Defense +1", 20, 0, 1),
           ("Defense +2", 40, 0, 2),
           ("Defense +3", 80, 0, 3),
           ("No Ring 1", 0, 0, 0),
           ("No Ring 2", 0, 0, 0)]

oponent = (104, 8, 1) # Hitpoints, Damage, Armor
me_hp = 100 
            
            
def fight(oponent, equipment):
    op_hp, op_dam, op_arm = oponent
    me_hp, me_dam, me_arm = equipment
    while True:
        op_hp = op_hp - me_dam + op_arm
        if op_hp <= 0:
            return True
        me_hp = me_hp - op_dam + me_arm
        if me_hp <= 0:
            return False
            
            
            
            
# Main Program
l_wins = []
for weapon in l_weapons:
    for armor in l_armor:
        for ring1, ring2 in list(combinations(l_rings, 2)):
            cost = weapon[1] + armor[1] + ring1[1] + ring2[1]
            damage = weapon[2] + ring1[2] + ring2[2]
            arm = armor[3] + ring1[3] + ring2[3]
            me = (me_hp, damage, arm)
            if fight(oponent, me):
                l_wins.append((cost, weapon[0], armor[0], ring1[0], ring2[0]))

min_cost = 1000
for win in l_wins:
    if win[0] <= min_cost:
        min_cost = win[0]
        min_win = win
print("Puzzle 1: Minimum costs to win:", min_win)

l_lost = []
for weapon in l_weapons:
    for armor in l_armor:
        for ring1, ring2 in list(combinations(l_rings, 2)):
            cost = weapon[1] + armor[1] + ring1[1] + ring2[1]
            damage = weapon[2] + ring1[2] + ring2[2]
            arm = armor[3] + ring1[3] + ring2[3]
            me = (me_hp, damage, arm)
            if not fight(oponent, me):
                l_lost.append((cost, weapon[0], armor[0], ring1[0], ring2[0]))

max_cost = 0
for lost in l_lost:
    if lost[0] >= max_cost:
        max_cost = lost[0]
        max_lost = lost
print("Puzzle 2: Maximum costs to loose:", max_lost)
                
            
```
