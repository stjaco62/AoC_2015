```python
# Advent of Code 2015, Day 18
def get_data(filename):
    l_lights = []
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            lights = ["."]
            for char in line:
                lights.append(char)
            lights.append(".")
            l_lights.append(lights)
        lights = []
        for i in range(len(l_lights[0])):
            lights.append(".")
        l_lights.append(lights)
        l_lights.insert(0, lights)
    return l_lights

def print_lights(l_lights):
    for lights in l_lights:
        for light in lights:
            print(light, end = "")
        print()
    print()
    
def calc_next_cell(l_lights, row, col):
    global life
    num_life = 0
    for i in range(-1, 2):
        for j in range(-1, 2):
            if l_lights[row + i][col + j] == "#":
                num_life += 1
    if l_lights[row][col] == "#":
        if int(life[0]) <= num_life - 1 <= int(life[1]):
            return ("#")
    else:
        if int(life[2]) <= num_life <= int(life[3]):
            return ("#")    
    return (".")
    
def calc_next_generation(l_lights):
    new_lights = []
    for i in range(len(l_lights)):
        row = []
        for j in range(len(l_lights[0])):
            row.append(".")
        new_lights.append(row)
    
    for i in range(1, len(l_lights) - 1):
        for j in range(1, len(l_lights[0]) - 1):
            new_lights[i][j] = calc_next_cell(l_lights, i, j)                
    return new_lights

def calc_lights(l_lights):
    tot = 0
    for lights in l_lights:
        tot += lights.count("#")
    return tot

def switch_on_corner(l_lights):
    l_lights[1][1] = "#"
    l_lights[1][len(l_lights[1]) - 2] = "#"
    l_lights[len(l_lights) - 2][1] = "#"
    l_lights[len(l_lights) - 2][len(l_lights[1]) - 2] = "#"
    return l_lights



# Main Program
life = "2333"
l_lights = get_data("Puzzle.txt")

for i in range(100):
    l_lights = calc_next_generation(l_lights)
    
tot = calc_lights(l_lights)
print("Puzzle 1: The total number of lights after 100 steps is", tot)

l_lights = get_data("Puzzle.txt")
l_lights = switch_on_corner(l_lights)

for i in range(100):
    l_lights = calc_next_generation(l_lights)
    l_lights = switch_on_corner(l_lights)
    #print_lights(l_lights)
    
tot = calc_lights(l_lights)
print("Puzzle 2: The total number of lights after 100 steps is", tot)


        
```
