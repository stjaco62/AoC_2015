```python
# Advent of Code 2015, Day 03
def get_data(filename):
    l_paths = []
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            l_paths.append(line)
    return l_paths

def track_path(path):
    x = 0
    y = 0
    d_pos = {}
    d_pos[(x, y)] = 1
    for char in path:
        if char == "^":
            y -= 1
        elif char == "v":
            y += 1
        elif char == "<":
            x -= 1
        elif char == ">":
            x += 1
        else:
            print("Error: Unknown Character:", char)
        if (x, y) in d_pos:
            d_pos[(x, y)] += 1
        else:
            d_pos[(x, y)] = 1
    return len(d_pos)

def track_path2(path):
    counter = 0
    x1 = 0
    y1 = 0
    x2 = 0
    y2 = 0
    d_pos = {}
    d_pos[(x1, y1)] = 1
    for char in path:
        if counter % 2 == 0:
            if char == "^":
                y1 -= 1
            elif char == "v":
                y1 += 1
            elif char == "<":
                x1 -= 1
            elif char == ">":
                x1 += 1
            else:
                print("Error: Unknown Character:", char)
            if (x1, y1) in d_pos:
                d_pos[(x1, y1)] += 1
            else:
                d_pos[(x1, y1)] = 1
        else:
            if char == "^":
                y2 -= 1
            elif char == "v":
                y2 += 1
            elif char == "<":
                x2 -= 1
            elif char == ">":
                x2 += 1
            else:
                print("Error: Unknown Character:", char)
            if (x2, y2) in d_pos:
                d_pos[(x2, y2)] += 1
            else:
                d_pos[(x2, y2)] = 1
        counter += 1
        
    return len(d_pos)

                
            





# Main Program
l_paths = get_data("Puzzle.txt")
for path in l_paths:
    #print(path)
    print(track_path(path), "houses are visited")
    print(track_path2(path), "houses are visited with robo-santa")
```
