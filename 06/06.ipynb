```python
# Advent of Code 2015, Day 06
def get_data(filename):
    l_instructions = []
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            if "toggle" in line:
                line = line[7:]
                cmd = "tgl"
            elif "turn on" in line:
                line = line[8:]
                cmd = "on"
            elif "turn off" in line:
                line = line[9:]
                cmd = "off"
            else:
                print("Error: Unknown command")
            nw, se = line.split(" through ")
            nw0, nw1 = nw.split(",")
            se0, se1 = se.split(",")
            nw0 = int(nw0)
            nw1 = int(nw1)
            se0 = int(se0)
            se1 = int(se1)
            l_instructions.append((cmd, nw0, nw1, se0, se1))
    return l_instructions

def init_lights(size):
    l_lights = []
    for i in range(size):
        row = []
        for j in range(size):
            row.append(False)
        l_lights.append(row)
    return l_lights

def init_lights_2(size):
    l_lights = []
    for i in range(size):
        row = []
        for j in range(size):
            row.append(0)
        l_lights.append(row)
    return l_lights


def count_lights(l_lights):
    count = 0
    for row in l_lights:
        for light in row:
            if light:
                count += 1
    return count

def exec_instruction(l_lights, instruction):
    cmd, nw0, nw1, se0, se1 = instruction
    for r in range(nw0, se0 + 1):
        for c in range(nw1, se1 + 1):
            if cmd == "tgl":
                l_lights[r][c] = not(l_lights[r][c])
            elif cmd == "off":
                l_lights[r][c] = False
            elif cmd == "on":
                l_lights[r][c] = True
    return l_lights

def calc_brightness(l_lights):
    brightness = 0
    for row in l_lights:
        for light in row:
            brightness += light
    return brightness

def exec_instruction_2(l_lights, instruction):
    cmd, nw0, nw1, se0, se1 = instruction
    for r in range(nw0, se0 + 1):
        for c in range(nw1, se1 + 1):
            if cmd == "tgl":
                l_lights[r][c] = l_lights[r][c] + 2
            elif cmd == "off":
                l_lights[r][c] = max(0, l_lights[r][c] - 1)
            elif cmd == "on":
                l_lights[r][c] = l_lights[r][c] + 1
    return l_lights

# Main Program
size = 1000
l_instructions = get_data("Puzzle.txt")
l_lights = init_lights(size)

for instruction in l_instructions:
    #print(instruction)
    l_lights = exec_instruction(l_lights, instruction)
    
num_lights = count_lights(l_lights)
print("Puzzle 1:", num_lights, "of", size**2, "lights are switched on")
            
    
# Puzzle 2 
l_lights = init_lights_2(size)

for instruction in l_instructions:
    #print(instruction)
    l_lights = exec_instruction_2(l_lights, instruction)
    
total_brightness = calc_brightness(l_lights)
print("Puzzle 2: Total brightness is:", total_brightness)
```
