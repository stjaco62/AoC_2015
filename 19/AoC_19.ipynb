```python
# Advent of Code 2015, Day 19, Problem 1
def get_data(filename):
    l_rules = []
    with open (filename, "r") as file:
        for line in file:
            line = line.strip()
            if line == "":
                pass
            elif not "=>" in line:
                pattern = line
            else:
                line = line.split(" => ")
                pat1 = line[0]
                pat2 = line[1]
                l_rules.append((pat1, pat2))
    return l_rules, pattern

def rep_one_occur(pattern, pat1, pat2):
    l_results = []
    for ind in range(len(pattern)):
        part1 = pattern[:ind]
        part2 = pattern[ind:]
        if pat1 in part2:
            part2 = part2.replace(pat1, pat2, 1)
            new_pattern = part1 + part2
            if not new_pattern in l_results:
                l_results.append(new_pattern)   
    return l_results

def recursive_replace(current_pattern, target_pattern, l_rules, depth):
    # Check break of recursion
    #print(current_pattern, depth)
    if current_pattern == target_pattern:
        #print("got it")
        return depth + 1
    if len(current_pattern) >= len(target_pattern):
        return -1

    # replace patterns and do recursion
    min_recursion = len(target_pattern) + 1
    for rule in l_rules:
        pat1, pat2 = rule
        l_results = rep_one_occur(current_pattern, pat1, pat2)
        #print(l_results)
        for result in l_results:
            num_recursion = recursive_replace(result, target_pattern, l_rules, depth + 1)
            if  num_recursion != -1:
                if num_recursion < min_recursion:
                    min_recursion = num_recursion
    if min_recursion == len(target_pattern):
        return -1
    else:
        return min_recursion + 1
    
# Main Program
l_rules, pattern = get_data("Puzzle.txt")
l_results = []

for rule in l_rules: 
    l_new_results = rep_one_occur(pattern, rule[0], rule[1])
    for result in l_new_results:
        if not result in l_results:
            l_results.append(result)
            
#for result in l_results:
#    print(result)

print("Puzzle 1: There are ", len(l_results), "solutions.")

#total = recursive_replace("e", pattern, l_rules, 0)
#print(total - 2)

# Recursion (forward) takes toooooo long :-(
```

```python
# Advent of Code 2015, Day 19, Problem 2, not deterministic, but works (based on comments in Reddit)
from random import shuffle

def get_data(filename):
    l_rules = []
    with open (filename, "r") as file:
        for line in file:
            line = line.strip()
            if line == "":
                pass
            elif not "=>" in line:
                pattern = line
            else:
                line = line.split(" => ")
                pat1 = line[0]
                pat2 = line[1]
                l_rules.append((pat1, pat2))
    return l_rules, pattern

def backward_replace(l_rules, pattern):
    counter = 0
    while pattern != "e":
        shuffle(l_rules)
        change = False
        for rule in l_rules:
            if rule[1] in pattern:
                pattern = pattern.replace(rule[1], rule[0], 1)
                counter += 1
                change = True
                break
        if not change:
            print("After", counter, "steps ended with pattern", pattern, "No more replacements possible.")
            return
    print("Done, ended with", pattern, "!!! Required", counter, "steps")


# Main Program
l_rules, pattern = get_data("Puzzle.txt")
for i in range(1, 30):
    backward_replace(l_rules, pattern)
```
